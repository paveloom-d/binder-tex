# Example

This example shows that TeX dependencies have been installed and can be used when creating plots using `matplotlib`.

Importing the packages:

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams as rcP
```

Generating data:

```python
x = np.arange(1000)
y = np.sin(np.pi * x / 100)
```

Customizing plots:

```python
# Set DPI
rcP["savefig.dpi"] = 300
rcP["figure.dpi"] = 300

# Activate TeX support
rcP["text.usetex"] = True

# Activate languages support
rcP["text.latex.preamble"] = r"\usepackage[main=english]{babel}"

# Switch to Computer Modern font set
rcP["mathtext.fontset"] = "cm"

# Set font sizes
rcP["font.size"] = 18
rcP["legend.fontsize"] = 12;
```

Creating a plot:

```python
plt.plot(x, y, color = "#425378")
plt.title(r"\textrm{A sample from} $\sin{\frac{\pi x}{100}}$", y = 1.02)
plt.xlabel(r"$x$")
plt.ylabel(r"$y$");
```
